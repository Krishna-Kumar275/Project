```python
Gender_encoded = pd.get_dummies(df_obj_col['Gender'],prefix= 'Gender',drop_first=True,dtype=int)

df_obj_col = df_obj_col.drop(columns=['Gender'])

df_obj_col = pd.concat([df_obj_col,Gender_encoded], axis= 1)
```

#  Scaling

```python
from sklearn.preprocessing import StandardScaler
```

```python
ss= StandardScaler()
```

```python
X_train_scale=pd.DataFrame(ss.fit_transform(X_train),columns=X_train.columns)
```

```python
X_test_scale=pd.DataFrame(ss.fit_transform(X_test),columns=X_test.columns)
```

```python
X_test_scale
```

```python
import tensorflow
```

```python
from tensorflow import keras
```

```python
from tensorflow.keras import Sequential
```

```python
from tensorflow.keras.layers import Dense
```

```python
model=Sequential()
```

```python
model.add(Dense(3,activation='sigmoid', input_dim=11))
```

```python
model.add(Dense(1,activation='sigmoid'))
```

```python
model.summary()
```

```python
model.compile(loss='binary_crossentropy',optimizer='adam')
```

```python
model.fit(X_train_scale,y_train,epochs=5)
```

```python
predicted_values=model.predict(X_test_scale)
```

```python
y_pred=np.where(predicted_values >0.5,1,0)
```

```python
y_pred
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy_score(y_test,y_pred)
```
